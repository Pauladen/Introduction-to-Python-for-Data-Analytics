### C. Hypothesis Testing

##### 7a. Statistically significant difference in sepal length among the different species:
#####    -One-Way ANOVA (Assuming data is Normal and Variances are homogenous- ie. equal)

```python
# Prepare data
x = iris[['SepalLengthCm']] # feature
y = iris['Species'] # target
# Run one-way ANOVA     from sklearn.feature_selection import f_classif
f_stat, p_value = f_classif(x, y)
# Print result
print(f'ANOVA F-statistics: {f_stat[0]:.4f}, \nP-value: {p_value[0]:.4f}')
# Interpretation
if p_value < 0.05:
    print("Result: there is a statistically significant difference in sepal length among the different species (reject the Ho)")
else:
    print("Result: No statistically significant difference in sepal length among the different species (fail to reject the Ho)")
```

##### 7b. Visualization: Boxplot for Spread and difference in Sepal length among the different Species

##### 8a. t-test to compare the means of two species for petal width

```python
# Select the two Species eg Iris-Setosa and Iris-versicolor
# Filter the DataFrame to include only these two Species

iris_Iris_Setosa = iris[iris['Species'] == 'Iris-setosa']
iris_Iris_Versicolor = iris[iris['Species'] == 'Iris-versicolor']
# Perform t-test
t_test, p_value = ttest_ind(iris_Iris_Setosa['PetalWidthCm'], iris_Iris_Versicolor['PetalWidthCm'])
print(f'T-Statistics: {t_test}, \nP-Value: {p_value}')
# Significance Level (alpha)
alpha = 0.05
# Interpretation
if p_value < alpha:
    print("Result: there is a statistically significant difference in Petal Width between Iris-setosa and Iris-versicolor Species (reject the Ho)")
else:
    print("Result: No statistically significant difference in Petal Width between Iris-setosa and Iris-versicolor Species (fail to reject the Ho)")
```

##### 10. Decision Boundaries Plot- for the classification of the species using a pair of features

```python
X = iris[['SepalLengthCm', 'SepalWidthCm']] # choosing pair of features
y = iris['Species']        # target
y_encoded = LabelEncoder().fit_transform(y)  # encoded label into numbers

model = SVC(kernel='linear') # Train a classifier (SVM with linear kernel)
model.fit(X, y_encoded)

# Prepare meshgrid
a_min, a_max = iris['SepalLengthCm'].min() - 0.5, iris['SepalLengthCm'].max() + 0.5
b_min, b_max = iris['SepalWidthCm'].min() - 0.5, iris['SepalWidthCm'].max() + 0.5
# Create a mesh grid using Numpy
aa, bb = np.meshgrid(np.linspace(a_min, a_max, 200), np.linspace(b_min, b_max, 200)) 
# Convert mesh grid to Pandas DataFrame for prediction
# Note the .T attribute which transposes the array. This is necessary because np.array creates a 2D with shape(2, n), 
# where n is the number of elements in aa and bb. By transposing the array, we get the desired shape (n, 2)
mesh_df = pd.DataFrame(np.array([aa.ravel(), bb.ravel()]).T, columns=['SepalLengthCm', 'SepalWidthCm']) 

# Predict class for each point in the mesh grid using trained model
Z = model.predict(mesh_df)  
Z = Z.reshape(aa.shape) 

plt.figure(figsize=(8, 6))
# Plot decision boundaries (decision regions) using contourf() to shade the regions 
plt.contourf(aa, bb, Z, alpha=0.3, cmap='coolwarm') 

sns.scatterplot(x=iris['SepalLengthCm'], y=iris['SepalWidthCm'], hue=iris['Species'], palette='coolwarm', edgecolor='k') # Scatter plot for better visualize

plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title(f'Decision Boundaries for Sepal Length and Sepal Width')
plt.legend(title='Species')
plt.show()
```

##### 12a. Identifying Outliers: Interquantile Range (IQR) Method (Statistical approach)

```python
Q1 = iris.iloc[:, 1:5].quantile(0.25)
Q3 = iris.iloc[:, 1:5].quantile(0.75)
IQR =Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
iris_outliers = ((iris.iloc[:, 1:5] < lower_bound) | (iris.iloc[:, [1, 2, 3, 4]] > upper_bound)) # Result display as Tuple of bool
print(iris_outliers.sum()) # Count of outliers in each column
# print(iris_outliers.count()) # Incorrect
print(iris_outliers)
```

##### 12a. Identifying Outliers: Box Plot Method (Visualization approach)

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris.iloc[:, 1:5]) 
plt.title("Boxplot of Iris Features to Detect Outliers")
plt.show()            # points outside the whiskers are potential outliers
```

##### 12b. Managing Outliers- Method 1: Remove Outliers

```python
iris_no_outliers1 = iris.iloc[:, 1:5][ ~ ((iris.iloc[:, 1:5] < (Q1 - 1.5 * IQR)) | (iris.iloc[:, 1:5] > (Q3 + 1.5 * IQR)))]
iris_no_outliers2 = iris.iloc[:, 1:5][ ~ iris_outliers]      # same as line above
print(iris_no_outliers1)
#print('\n', iris_no_outliers2) # Same as line above
print('\n', iris.merge(iris_no_outliers1)) # what does the merge() method does? 48 rows seems to be missing after merge.
```

##### 12b. Managing Outliers- Method 2: Replace Outlier with Median

```python
iris_no_outliers = iris.iloc[:, 1:5].copy()    # Create a copy of the original DataFrame to be modified

for col in iris_no_outliers:
    median_value = iris_no_outliers[col].median()
    iris_no_outliers[col] = iris_no_outliers[col].mask(((iris_no_outliers < (Q1 - 1.5 * IQR)) | (iris_no_outliers > (Q3 + 1.5 * IQR)))[col], median_value)
   # iris_no_outliers[col] = iris_no_outliers[col].mask(iris_outliers[col], median_value) # Same as above line
print(iris_no_outliers.head())  # Outliers replaced with median
```

### Classification & Predictions

##### 15a. Split the dataset into training & testing sets using Scikit-learn (**see 15b**)

##### 15b. Prepare iris dataset for machine learning

```python
print(iris.iloc[:, 1:5])
```

```python
# Check for missing values
iris_isnull = iris.isnull().sum()
print(iris_isnull)

# Encode categorical variables

le = LabelEncoder()                 # Creates an instance of LabelEncoder
iris['species_encoded'] = le.fit_transform(iris['Species'])

# Split the dataset into features and encoded target
X = iris.iloc[:, 1:5]      # features (SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm)
y = iris['species_encoded']      # encoded target


# Split the dataset into training (70%) and testing (30%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)       # using train_test_split of sklearn

# Scale the features using StandardScaler                # This preprocessing step was NOT done in splitting dataset to training & testing sets above

scaler = StandardScaler()                     # Creates an instance of StandardScaler
X_train_scaled = scaler.fit_transform(X_train)         # Fit on training data, then transform
X_test_scaled = scaler.transform(X_test)               # Transform test data using the same scaler
# Now your iris dataset is ready for machine learning modeling

print('\n', X_train_scaled) 
print('\n', X_test_scaled)
```

##### 16. Analysis: feature importance; Model: Random Forest for classification

```python
# Continuation of code snippets on preparing iris dataset for machine learning

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)   # Initialize model (create instance)
rf_model.fit(X_train, y_train)                                         # Train model
feature_importance = rf_model.feature_importances_               # extract feature importance from model's attribute (store)
print(feature_importance)
iris_importance = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importance})    # Create DataFrame for better visualization
iris_importance = iris_importance.sort_values(by='Importance', ascending=False)
print()
print(iris_importance)
plt.figure(figsize=(8, 5))                                        # Visualize feature importance on Horizontal Bar Plot
plt.barh(iris_importance['Feature'], iris_importance['Importance'], color='skyblue')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.title('Feature Importance in Random Forest Model')
plt.gca().invert_yaxis()  # Invert y-axis to have the most important feature on top
plt.show()
```
