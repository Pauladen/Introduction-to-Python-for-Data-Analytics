```python
%matplotlib inline
import matplotlib.pyplot as plt
from itertools import combinations
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
for x_index, y_index in combinations(range(4), 2):
    plt.scatter(iris.data[:, x_index], iris.data[:, y_index],c=iris.target)
    # this formatter will label the colorbar with the correct target names
    formatter = plt.FuncFormatter(lambda i, *args:
                                  iris.target_names[int(i)])
    plt.colorbar(ticks=[0, 1, 2], format=formatter)
    plt.xlabel(iris.feature_names[x_index])
    plt.ylabel(iris.feature_names[y_index])
    plt.show()
```

```python
# As seen in the figures above, a combination of two parameters which maximally separate the three classes is:
# petal length (cm) petal width (cm), that is, (2, 3)
```

```python
from sklearn.datasets import fetch_olivetti_faces
```

```python
# fetch the faces data. This takes some time.
faces = fetch_olivetti_faces()
print(faces)
print('\n', faces.data.shape)
print('\n', faces.target.shape)
print('\n', faces.images.shape)
```

```python
images = faces.images

# hint: plt.cm.bone is a good colormap for this data
# set up the figure
fig = plt.figure(figsize=(6, 6))  # figure size in inches
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# plot the digits: each image is 8x8 pixels
for i in range(100):
    ax = fig.add_subplot(10, 10, i + 1, xticks=[], yticks=[])
    ax.imshow(images[i], cmap=plt.cm.bone, interpolation='nearest')
    
    # label the image with the target value
    ax.text(0, 7, str(faces.target[i]))
plt.show()
```

```python
images = faces.images
unique_ids = np.unique(faces.target)

# creating 4x10 subplots in 18x9 figure size
fig, axarr = plt.subplots(nrows=4, ncols=10, figsize=(18, 9))
# for easy iteration, flatten 4x10 subplots matrix to 40 array
axarr = axarr.flatten()
#iterating over user ids
for uniq_id in unique_ids:
    image_index = uniq_id * 10
    axarr[uniq_id].imshow(images[image_index], cmap='gray')
    axarr[uniq_id].set_xticks([])
    axarr[uniq_id].set_yticks([])
    axarr[uniq_id].set_title('face id:{}'.format(uniq_id))
plt.suptitle('There are 40 distinct people in the dataset')
plt.show()
```

```python
# Define subject IDs
subject_ids = [0, 1, 2, 3]
cols = 10  
rows = int(np.ceil(len(subject_ids) * 10 / cols))

fig, axarray = plt.subplots(nrows=rows, ncols=cols, figsize=(18, 9))

for i, subject_id in enumerate(subject_ids):
    for j in range(cols):
        image_index = subject_id * 10 + j
        axarray[i, j].imshow(images[image_index], cmap='gray')
        axarray[i, j].set_xticks([])
        axarray[i, j].set_yticks([])
        axarray[i, j].set_title(f'face id: {subject_id}')
plt.show()
```
