##LINE PLOTS USING MATPLOTLIB

```python
# Line plot.
# Importing matplotlib to plot the graphs.
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Importing pandas for using pandas dataframes.
import pandas as pd
import gdown
```

```python
# Replace with your Google Drive file ID
file_id = '1nfqjxGIpLEXCJGhAnsgHrt9n6WK9ZtR8'

# Download the file to your local environment
url = f'https://drive.google.com/uc?id={file_id}'
output = 'property_tax_report_2018.csv'  # Choose a local file name
gdown.download(url, output, quiet=False)


# Now read the downloaded CSV file
df = pd.read_csv(output)
df.head()
```

```python
# Removing the null values in PROPERTY_POSTAL_CODE.
df = df[(df['PROPERTY_POSTAL_CODE'].notnull())]
```

```python
# Grouping by YEAR_BUILT and aggregating based on PID to count the number of properties for each year.
df = df[['PID', 'YEAR_BUILT']].groupby('YEAR_BUILT', as_index = False).count().astype('int').rename(columns = {'PID':'No_of_properties_built'})
```

```python
# Filtering YEAR_BUILT and keeping only the values between 1900 to 2018.
df = df[(df['YEAR_BUILT'] >= 1900) & (df['YEAR_BUILT'] <= 2018)]
```

```python
# X-axis: YEAR_BUILT
x = df['YEAR_BUILT']
```

```python
# Y-axis: Number of properties built.
y = df['No_of_properties_built']
```

```python
# Change the size of the figure (in inches).
plt.figure(figsize=(17,6))
```

```python
# Plotting the graph using x and y with 'dodgerblue' color.
# Different labels can be given to different lines in the same plot.
# Linewidth determines the width of the line.
plt.plot(x, y, 'dodgerblue', label = 'Number of properties built', linewidth = 1)
# plt.plot(x2, y2, 'red', label = 'Line 2', linewidth = 2)

# X-axis label.
plt.xlabel('YEAR', fontsize = 16)

# Y-axis label.
plt.ylabel('Number of properties built', fontsize = 16)

# Title of the plot.
plt.title('Number of houses built between\n1900 and 2018', fontsize = 16)

# Grid
# plt.grid(True)
plt.grid(False)

# Legend for the plot.
plt.legend()

# Saving the figure on disk. 'dpi' and 'quality' can be adjusted according to the required image quality.
plt.savefig('Line_plot.jpeg', dpi = 750)

# Displays the plot.
plt.show()
```

```python
# Reading the input file.
file_path = "/content/property_tax_report_2018.csv"
dff = pd.read_csv(file_path)
dff.head()
```

```python
# Get column names
print(dff.columns)
```

```python
# Reading the input file.
#dff = pd.read_csv("property_tax_report_2018.csv")
```

```python
# Filtering out the null values in ZONE_CATEGORY
dff = dff[dff['ZONING_CLASSIFICATION'].notnull()]
```

```python
# Grouping by ZONE_CATEGORY and aggregating based on PID to count the number of properties for a particular zone.
dff_zone_properties = dff.groupby('ZONING_CLASSIFICATION', as_index=False)['PID'].count().rename(columns = {'PID':'No_of_properties'})
```

```python
# Counting the total number of properties.
total_properties = dff_zone_properties['No_of_properties'].sum()
```

```python
# Calculating the percentage share of each ZONE for the number of properties in the total number of properties.
dff_zone_properties['percentage_of_properties'] = ((dff_zone_properties['No_of_properties'] / total_properties) * 100)
```

```python
# Finding the ZONES with the top-6 percentage share in the total number of properties.
dff_top_6_zone_percentage = dff_zone_properties.nlargest(columns='percentage_of_properties', n = 6)
```

```python
# Change the size of the figure (in inches).
plt.figure(figsize=(8,6))

# Slices: percentage_of_properties.
slices = dff_top_6_zone_percentage['percentage_of_properties']
# Categories: ZONE_CATEGORY.
categories = dff_top_6_zone_percentage['ZONING_CLASSIFICATION']
# For different colors: https://matplotlib.org/examples/color/named_colors.html
cols = ['purple', 'red', 'green', 'orange', 'dodgerblue', 'yellow']

# Plotting the pie-chart.
plt.pie(slices,
        labels = categories,
        colors=cols,
        startangle = 90,
#         shadow = True,
        # To drag the slices away from the centre of the plot.
        explode = (0.040,0.030,0.020,0.010,0.030,0.040),
        # To display the percent value using Python string formatting.
        autopct = '%1.1f%%'
       )

# Title of the plot.
plt.title('Top-10 zone categories with the highest percentage\nshare in the total number of properties.', fontsize = 12)

# Saving the figure on disk. 'dpi' and 'quality' can be adjusted according to the required image quality.
plt.savefig('Pie_chart.jpeg', dpi = 750)

# Displays the plot.
plt.show()
```

## HISTOGRAM PLOT USING MATPLOTLIB

```python
# Removing the null values in PROPERTY_POSTAL_CODE.
df = df[(df['PROPERTY_POSTAL_CODE'].notnull())]
```

```python
# Grouping by YEAR_BUILT and aggregating based on PID to count the number of properties for each year.
df = df[['PID', 'YEAR_BUILT']].groupby('YEAR_BUILT', as_index = False).count().astype('int').rename(columns = {'PID':'No_of_properties_built'})
```

```python
# Filtering YEAR_BUILT and keeping only the values between 1900 to 2018.
df = df[(df['YEAR_BUILT'] >= 1900) & (df['YEAR_BUILT'] <= 2018)]
```

```python
# Change the size of the figure (in inches).
plt.figure(figsize=(17,6))

# X-axis: Number of properties built from 1900 to 2018.
# Y-axis: Frequency.
plt.hist(df['No_of_properties_built'],
         bins = 50,
         histtype='bar',
         rwidth = 1.0,
         color = 'dodgerblue',
         alpha = 0.8
       )

# X-axis label.
plt.xlabel('Number of properties built from 1900 to 2018', fontsize = 16)

# Y-axis label.
plt.ylabel('Frequency', fontsize = 16)

# Title of the plot.
plt.title('Distribution of the number of properties built between 1900 and 2018.', fontsize = 16)

# Grid
# plt.grid(True)
plt.grid(axis='y')

# Saving the figure on disk. 'dpi' and 'quality' can be adjusted according to the required image quality.
plt.savefig('Histogram.jpeg', dpi = 750)

# Displays the plot.
plt.show()
```
